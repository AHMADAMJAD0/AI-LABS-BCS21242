```python

# 9 write a python program to get the fibonacci srires between 0 to 50 -------------
def fibonacci(n):
    fib_sequence = [0, 1]
    while fib_sequence[-1] < n:
        fib_sequence.append(fib_sequence[-1] + fib_sequence[-2])
    return fib_sequence

fib_sequence = fibonacci(50)
print(fib_sequence)
```

```python
 # 11 

def create_table(i, j):
    rows = i
    columns = j
    arr = []
    
    for i in range(rows):
        row = []
        for j in range(columns):
            row.append(i * j)
        arr.append(row)
        
    return arr

def print_table(table):
    for row in table:
        for cell in row:
            print(cell, end=' ')
        print()

def main():
    # Taking user input for rows and columns
    rows = int(input("Enter the number of rows: "))
    columns = int(input("Enter the number of columns: "))
    
    table = create_table(rows, columns)
    print_table(table)
    
main()
```

```python
# 13 


def main():
    binary_numbers = input("Enter a sequence of comma separated 4 digit binary numbers: ")
    binary_numbers = binary_numbers.split(',')
    
    for binary_number in binary_numbers:
        decimal_number = int(binary_number, 2)
        
        if decimal_number % 5 == 0:
            print(binary_number, end=', ')

main()
```

```python
 # 14


def main():
    string = input("Enter a string: ")
    
    letters = 0
    digits = 0
    
    for char in string:
        if char.isalpha():
            letters += 1
        elif char.isdigit():
            digits += 1
    
    print(f"Letters {letters}")
    print(f"Digits {digits}")
    
main()
```

```python
# 15

def main():
    password = input("Enter a password: ")
    
    if len(password) < 6 or len(password) > 16:
        print("Password must be between 6 and 16 characters")
        return
    
    has_lower = False
    has_upper = False
    has_digit = False
    has_special = False
    
    for char in password:
        if char.islower():
            has_lower = True
        elif char.isupper():
            has_upper = True
        elif char.isdigit():
            has_digit = True
        elif char in "$#@":
            has_special = True
    
    if has_lower and has_upper and has_digit and has_special:
        print("Valid password")
    else:
        print("Invalid password")
        
main()
```
